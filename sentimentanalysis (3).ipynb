## Calculate lexical diversity for tweets

```python
#Lexical diversity is the ratio of number of datasets used in the data
# Assumption is that there si no repetition of the same words
'''

The functions 
'''
def calculate_lexical_diversity(column_name):
    """Calculates lexical diversity for a given column."""
    tokens = [word for sublist in df[column_name].astype(str).str.split() for word in sublist]
    return len(set(tokens)) / len(tokens)

def calculate_average_words(column_name):
    """Calculates average words for a given column."""
    word_counts = merged_df[column_name].astype(str).str.len()
    return word_counts.mean()

# Calculate lexical diversity and average words for each column
lexical_diversity_likes = calculate_lexical_diversity('likes')
lexical_diversity_quotes = calculate_lexical_diversity('quotes')
lexical_diversity_retweets = calculate_lexical_diversity('retweets')
lexical_diversity_comments = calculate_lexical_diversity('comments')

average_words_likes = calculate_average_words('likes')
average_words_quotes = calculate_average_words('quotes')
average_words_retweets = calculate_average_words('retweets')
average_words_comments = calculate_average_words('comments')

# Print the results
print("Lexical Diversity:")
print("Likes:", lexical_diversity_likes)
print("Quotes:", lexical_diversity_quotes)
print("Retweets:", lexical_diversity_retweets)
print("Comments:", lexical_diversity_comments)

print("\nAverage Words:")
print("Likes:", average_words_likes)
print("Quotes:", average_words_quotes)
print("Retweets:", average_words_retweets)
print("Comments:", average_words_comments)
```

```python
from prettytable import PrettyTable

retweets = [
    (
        status['retweet_count'],
        status['retweeted_status']['user']['screen_name'],
        status['retweeted_status']['retweets'],
        status['text']
    )
    for status in merged_df.to_dict('records')
    if 'retweeted_status' in status
]

# Slice off the first 5 from the results
top_retweets = sorted(retweets, reverse=True)[:5]

# Create and configure the PrettyTable instance
pt = PrettyTable(field_names=[ "user","likes", "retweets", "comments"])

# Set alignment for all columns (you can set individual alignments too)
pt.align = "l"  # Left-align all columns

# Add rows to the table
for row in top_retweets:
    pt.add_row(row)

# Set maximum width for the "Text" column
pt.max_width["Text"] = 50

# Print the table
print(pt)
```
